```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')


import pylab as plt
import seaborn as sns

from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By   # para buscar segun tag, id, class, etc...
from selenium import webdriver
import time

from selenium import webdriver
```

```python
opciones=Options()

opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless=False    # si True, no aperece la ventana (headless=no visible)
opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
opciones.add_argument('--incognito')
```

```python
PATH=ChromeDriverManager().install()
```

```python
url = 'https://www.nba.com/stats/teams/traditional?Season=2021-22&SeasonType=Playoffs&PerMode=PerGame&PORound=1'
```

```python
driver = webdriver.Chrome(PATH)

driver.get(url)
```

```python
tabla=driver.find_element(By.CLASS_NAME, 'Crom_body__UYOcU')

teams = [[e.text for e in f.find_elements(By.TAG_NAME, 'td')] 
      for f in tabla.find_elements(By.TAG_NAME, 'tr')]
```

```python
cabeceras = driver.find_element(By.CLASS_NAME, 'Crom_headers__mzI_m')

cabeceras = [c.text for c in cabeceras.find_elements(By.TAG_NAME, 'th')]

headings = cabeceras[0:28]
```

```python
semis_2021 = pd.DataFrame(teams, columns = headings)
semis_2021
```

```python
semis_2021['SEASON'] = '2022'
```
